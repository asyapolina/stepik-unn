# 1.1
Вы сталкивались с хитрыми математическими фокусами, позволяющими отгадать Ваш возраст с помощью последовательности математических операций? Сейчас на Ваших глазах произойдёт магия. Вы помните свой размер обуви? Припишите к этому числу два ноля. Теперь из полученного результата необходимо вычесть Ваш год рождения. Затем прибавьте к полученному числу текущий год. Посмотрите на две последние цифры результирующего числа - это и есть Ваш возраст (сколько лет Вам уже исполнилось или исполнится в текущем году). Удивлены? 

А теперь попробуйте написать программу, на вход которой даётся три числа через пробел - размер обуви, год рождения и текущий год. Основываясь на предложенном алгоритме, программа должна вывести возраст человека.

```python
a, b, c = map(int, input().split(" "))
age = (a * 100 - b + c) % 100
print(age)
```

# 1.3
В отделении банка в A1 часов B1 минут C1 секунд чихнул клиент, заражённый гриппом. В момент времени A2 часов B2 минут C2 секунд тех же суток в этом банке появился другой посетитель. Чтобы определить вероятность заражения, необходимо рассчитать количество секунд между моментами времени, когда в банк зашли второй и первый клиент.

Входные данные программы представляют собой три целых числа в первой строке через двоеточие: часы, минуты, секунды, задающие момент времени прихода первого клиента и три целых числа во второй строке, задающих момент времени прихода второго клиента.

Выходные данные должны содержать число секунд между этими моментами времени.

```python
A1, B1, C1 = map(int, input().split(":"))
A2, B2, C2 = map(int, input().split(":"))
sec1 = A1 * 3600 + B1 * 60 + C1 
sec2 = A2 * 3600 + B2 * 60 + C2
print(sec2 - sec1)
```

# 1.4
Васе предстоит сдать сложный экзамен по биологии. Количество экзаменационных билетов составляет bb штук (bb - чётное число). К сожалению, после предыдущего экзамена оставалось 2 дня на подготовку к биологии, поэтому Вася успел выучить только половину билетов. Утром в день экзамена он выучил ещё один билет. Какова вероятность того, что Васе на экзамене достанется выученный билет?

Входные данные: количество экзаменационных билетов bb.

Выходные данные: вероятность достать выученный билет.

Ответ вашей программы считается корректным, если он отличается от правильного ответа не более чем на 10^-6.

```python
b = int(input())
p = (b / 2 + 1) / b
print(p)
```

# 1.5 
Пётр Васечкин долго мечтал и наконец-то купил квартиру. К сожалению, его накоплений хватило только на квартиру в панельном доме со старым ремонтом. С наступлением зимы он обнаружил, что в квартире очень холодно и вознамерился заменить старые радиаторы отопления. Выбор его пал на биметаллические радиаторы фирмы «Супер-батареи», которые представляют собой совокупность секций, соединённых между собой посредством резьбового соединения. 

Перед покупкой закономерно возник вопрос: «Скольки секционный радиатор необходимо купить, чтобы поддерживать оптимальную температуру в помещении?» Поискав в интернете Пётр нашёл следующую формулу для расчёта количества секций $K~=~{объём \: помещения \: \cdot \: норма \: затрат \: тепла \over теплоотдача \: одной \: секции} $
​
 . Он измерил длину l, ширину w и высоту h комнаты в метрах, узнал, что норма затрат тепла для обогрева одного кубометра помещений в панельном доме составляет 41 Вт/м^3, а теплоотдача одной секции «Супер-батареи» равна t Вт . Вам нужно помочь Петру и рассчитать количество секций с округлением в большую сторону.

Входные данные представляют собой строку, содержащую следующие величины через пробел: длина l, ширина w, высота h комнаты и теплоотдача одной секции «Супер-батареи» t.

Выходные данные должны содержать количество секций K.

```python
import math
l, w, h, t = map(float, input().split(" "))
k = (l * w * h * 41) / t
print(math.ceil(k))
```
