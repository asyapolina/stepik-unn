# 3.1
На вход программе последовательно подаются ненулевые целые числа (каждое число в новой строке). Ввод чисел заканчивается при вводе 0. Необходимо определить сумму всех положительных и сумму всех отрицательных введённых элементов. Программа должна вывести два значения - сумму положительных (на первом месте) и сумму отрицательных элементов (на втором месте) через запятую. Если сразу был введён ноль, то обе суммы равны нулю.

```python
n = int(input())
sumn1 = 0
sumn2 = 0
while n != 0:
    if n > 0:
        sumn1 += n
    if n < 0:
        sumn2 += n
    n = int(input()) 
print(f"{sumn1},{sumn2}")
```

# 3.2 
Алина работает администратором в спортивном зале. В спортзал ежедневно приходит некоторое количество посетителей, N. Алине поручили вести статистику и в конце каждого рабочего дня определять возраста самого молодого и самого старшего из посетителей. Чтобы Алина не выполняла данную задачу вручную, необходимо написать программу, которая сделает за неё эту работу.

На вход программе подаётся строка, содержащая N+1 число, из которых первое число - это количество посетителей NN спортзала за день, а последующие числа - это возраста посетителей. Числа задаются через пробел.

Основываясь на входной информации, программа должна определить (не используя специальные встроенные функции языка программирования) возраста самого молодого и самого старшего посетителей и вывести их в порядке возрастания, разделив пробелом. Если посетителей не было (количество посетителей равно нулю), то вывести нулевые значения.

```python
N = list(map(int, input().split()))
if N[0] > 0:
    maxN = N[1]
    minN = N[1]
    for i in range(1, len(N)):
        if N[i] > maxN:
            maxN = N[i]
        elif N[i] < minN:
            minN = N[i]
    print(minN, maxN)
else: 
    print(0, 0)
```

# 3.3
В клинике «Семейная клиника» проводится акция на УЗИ беременных женщин. Стажёру Анне поручено позвонить всем беременным женщинам, наблюдающимся в клинике, и сообщить об акции. Так как число пациентов клиники NN может быть очень большим, необходимо помочь Анне автоматизировать процесс выбора беременных женщин среди всех пациентов.

На вход программе подаются следующие входные данные: в первой строке - число N всех пациентов клиники, далее идёт ровно N строк, содержащих следующую информацию через пробел - порядковый номер пациента (1, 2, 3,...), его пол (M - мужчины, F - женщины) и наличие беременности (1 - беременна, 0 - не беременна).

Программа должна вывести через запятую порядковые номера всех беременных женщин. Если беременных нет, то вывести 0.

```python
n = int(input())
res = []
for i in range(n):
    lst = list(input().split())
    if lst[1] == "F" and lst[2] == "1":
        res.append(lst[0])
if len(res) == 0:
    print("0")
else:
    print(*res, sep=",")
```

# 3.4
Хорошо известно, что в производственной лаборатории должны контролироваться температура и относительная влажность воздуха. Для измерения климатических параметров лаборатории часто используются термогигрометры, а результаты измерений заносятся в специальный журнал. Обычно температура в лаборатории колеблется от 15 до 27 градусов, а влажность воздуха изменяется от 15 до 75 процентов. Необходимо на основании всех сделанных за день записей определить среднесуточную температуру и среднесуточную влажность воздуха в лаборатории. В течение каждого дня может быть сделано разное количество записей, но всегда есть хотя бы одна запись температуры и хотя бы одна запись влажности.

На вход программе через пробел подаются зарегистрированные записи климатических измерений. Пример записи температуры: t=21C, влажности: h=52%. Все значения температур и влажностей заранее округлены до целых значений.

Необходимо вычислить среднесуточную температуру N и среднесуточную влажность воздуха M и вывести результат в формате T=N;H=M (без пробелов).

```python
lst = list(input().split(" "))
temp = []
hum = []
for elem in lst:
    if "t" in elem:
        temp.append(int(elem[2:4]))
    elif "h" in elem:
        hum.append(int(elem[2:4]))
print(f"T={sum(temp)/len(temp):.2f};H={sum(hum)/len(hum):.2f}")
```

# 3.5
Напишите программу, реализующую алгоритм перевода целого положительного числа N из двоичной системы счисления в десятичную.

На вход программе подаётся число NN, представленное в двоичной системе счисления.

Программа должна вывести число NN, представленное в десятичной системе счисления.

```python
n = input()
l = len(n)
res = 0
for i in range(l):
    res += int(n[i]) * 2 ** (l - i - 1)
print(res)
```
