# 4.1

Одна из цепочек молекулы ДНК имеет порядок нуклеотидов, записанный в строке S, которая подаётся на вход программе.

Необходимо вывести на экран порядок нуклеотидов в комплементарной цепочке ДНК.

```python
def complementary_seq(seq):
    mapped_nucl = {"A":"T", "T":"A", "G":"C", "C":"G"}
    comp_seq = ""
    for nucl in seq:
        comp_seq += mapped_nucl[nucl]
    return comp_seq

dna = input()
print(complementary_seq(dna))
```

# 4.2
Необходимо определить, является ли введённая строка палиндромом. Вывести yes, если является, и no в противном случае.

```python
s = input()
if s == s[::-1]:
    print("yes")
else:
    print("no")
```

```python
s = input()
flag = True
for i in range(len(s) // 2):
    if s[i] != s[len(s) - 1 - i]:
        flag = False
        break
if flag:
    print("yes")
else:
    print("no")
```

# 4.3
Необходимо вычислить расстояние Хэмминга для двух нуклеотидных последовательностей.

Входные данные: две строки, в каждой из которых записана последовательность нуклеотидов одинаковой длины.

Выходные данные: расстояние Хэмминга.


```python
s1 = input()
s2 = input()
dist_h = 0
if len(s1) == len(s2):
    for i in range(len(s1)):
        if s1[i] != s2[i]:
            dist_h += 1
    print(dist_h)
else:
    print("ошибка")
```

# 4.4
На вход программе подаётся фрагмент нуклеотидной последовательности молекулы ДНК. Необходимо определить нуклеотидную последовательность второй цепи и общее число водородных связей, которые образуются между двумя цепями ДНК. 

Программа должна вывести число водородных связей.

```python
dna = input()
dict_dna = {"A":"T", "T":"A", "G":"C", "C":"G"}
dna_new = ""
h_bonds = 0
for nucl in dna:
    dna_new += dict_dna[nucl]
    if nucl == "A" or nucl == "T":
        h_bonds += 2
    elif nucl == "C" or nucl == "G":
        h_bonds += 3
print(h_bonds)
```

# 4.5
Одно секретное предприятие для переписки между сотрудниками придумало специальный шифр. Каждая буква зашифрованного слова после её встречи в тексте может несколько раз (или ни разу) повторяться в случайных местах слова, однако, символ ";" указывает на то, что повторения всех предыдущих букв дальше не будет. Символ ";" может встречаться неопределённое количество раз в начале, в конце или внутри слова. В слове могут встречаться только буквы русского алфавита и символ ";".

Например, по секретному каналу было передано слово сююссрюсрсссрпрю;ррририззиррз. Как его расшифровать? Буква с встречается впервые, значит, она является первой буквой искомого слова; буква ю так же встречается впервые, поэтому является второй буквой искомого слова; буквы юсс являются повторениями равнее встреченных букв с и ю, поэтому они игнорируются; буква р встречается впервые и добавляется к искомому слову; последующие буквы юсрссср являются повторениями ранее встречающихся букв, поэтому игнорируются; буква п встречается впервые и добавляется к искомому слову; буквы рю игнорируются по понятным причинам. Далее появляется символ ; , который означает, что дальнейшее появление букв с, ю, р, п не будет расцениваться как повторения, и эти буквы вновь могут участвовать в словообразовании. После ; идёт буква р, которая добавляется к исходному слову, а вот уже дальнейшее появление буквы р будет расцениваться как повторение. Доводя процедуру декодировки до конца, получаем, что в искомом слове так же присутствуют буквы и, з, а само слово - сюрприз.

Необходимо написать программу-декодировщик, на вход которой подаётся закодированное слово, а на выходе - изначальная запись слова. Слова задаются на русском языке строчными буквами.


```python
s = input().split(";") # разделяем строку на слова, получаем список?
m = []
for fragment in s: # по слову в списке
    filtered_s = ""
    #получаем 1 фрагмент из зашифрованного слова
    for ch in fragment: # по буквам в словах 
        if ch not in filtered_s:
            filtered_s += ch # получили 1 символ 
    m.append(filtered_s)   
print("".join(m))
```
