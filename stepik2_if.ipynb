# 2.1
Максим считает, что обладает экстрасенсорными способностями и утверждает, что может угадать любое целое число от 0 до 100, которое загадает любой другой человек. Тогда друзья решили воспользоваться специальной компьютерной программой. Максим выходит в соседнюю комнату, пока Андрей вводит загаданное число. Затем возвращается Максим и вводит число, которое по его мнению ввёл Андрей. Программа должна определить, угадал ли Максим загаданное число.

На вход программе в первой строке подаётся число, которое загадал Андрей, а во второй строке - число, которое напечатал Максим.

Программа должна вывести на экран yes, если Максим угадал число, и no в противном случае.

```python
n1 = int(input())
n2 = int(input())
if n1 == n2:
    print("yes")
else:
    print("no")
```

# 2.3 
Врач прописал Марии Семёновне принимать некий препарат по одной таблетке X раз в день
 
$(1≤X≤4)$ в течение Y недель $(1≤Y≤6)$. Производитель выпускает данное лекарство в упаковках по 20, 30, 50 и 100 таблеток. В аптеке, в которую ходит Мария Семёновна, не всегда есть в наличии большие объёмы нужного лекарства, поэтому женщина иногда не может купить требуемое на весь курс количество таблеток. Известно, что она купила K упаковок $(1≤K≤10)$ по N таблеток. Необходимо выяснить, хватит ли купленного количества таблеток на весь курс и придётся ли Марии Семёновне докупать лекарство.

На вход программе подаются значения X, Y, K, N записанные в строке через запятую.

Программа должна вывести на экран enough, в случае, если купленного лекарства достаточно на курс или not enough в противном случае.

```python
x, y, k, n = map(int, input().split(","))
z = x * 7 * y
p = k * n 
if p >= z:
    print("enough")
else:
    print("not enough")
```

# 2.4 
Даны три целых числа X, Y, Z, записанных в строку через пробел. Необходимо определить, существует ли треугольник с такими длинами сторон и, если существует, будет ли он прямоугольным.

Возможные варианты вывода программы: "no" (треугольник не существует), "yes-yes" (треугольник существует и он прямоугольный), "yes-no" (треугольник существует, но он не является прямоугольным).

```python
x, y, z = map(int, input().split(" "))
if (x < y + z) and (y < x + z) and (z < y + x):
    if (x ** 2 == z ** 2 + y ** 2) or (z ** 2 == x ** 2 + y ** 2) or (y ** 2 == z ** 2 + x ** 2):
        print("yes-yes")
    else:
        print("yes-no")
else:
    print("no")
```

# 2.5 
Михаил работает в ботаническом саду - ухаживает за растениями. В назначенный день в сад привезли M литров жидкого удобрения, но новый поставщик поменял условия доставки: теперь нужно сразу же переливать удобрение в свои ёмкости, а тару возвращать курьеру. В спешке Михаил нашёл две ёмкости: кубическую с ребром AA метров и цилиндрическую c высотой H метров и радиусом основания R метров. Помогите Михаилу быстро определить, в какую ёмкость поместится удобрение.

На вход программе подаются 4 вещественных числа: M, A, H, R через пробел.

Необходимо вывести на экран: cube, если удобрение целиком поместится в кубическую ёмкость (а в цилиндрическую не поместится); cylinder, если удобрение целиком поместится в цилиндрическую ёмкость (а в кубическую не поместится); cube-or-cylinder, если удобрение поместится как в кубическую, так и в цилиндрическую ёмкости; cube-and-cylinder, если удобрение не помещается целиком ни в одну из ёмкостей, но может быть разлито сразу по двум ёмкостям; fail, если этих ёмкостей не достаточно, чтобы целиком разлить удобрение.

```python
M, A, H, R = map(float,input().split(" "))
V_cyl = 3.14 * H * R ** 2 * 1000
V_cube = A ** 3 * 1000
if M < V_cube and M < V_cyl:
    print("cube-or-cylinder") 
elif M < V_cube and M > V_cyl:
    print("cube")
elif M > V_cube and M < V_cyl:
    print("cylinder")
elif M < V_cube + V_cyl:
    print("cube-and-cylinder")
else:
    print("fail")
```
